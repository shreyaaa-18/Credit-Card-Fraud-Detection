<h1>Credit Card Fraud Detection</h1>

<b>Importing the dependencies</b>

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a Pandas DataFrame 
credit_card_data = pd.read_csv('Downloads/Practice/Credit Card Fraud Detection/creditcard.csv')
```

```python
# first 5 rows of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset information 
credit_card_data.info()
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

<b>This Dataset is highly unbalanced</b>

0 -> Normal Transaction <br/>
1 -> Fraudulent Transaction

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

<b>Under-Sampling<b/>

Build a sample dataset containing distrubution of normal transactions and fraudulent transactions </br></br>
Number of Fraudulent Transaction -> 492

```python
legit_sample = legit.sample(n=492)
```

Concatenationg two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

<b>Splitting the data into Features & Targets</b>

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

<b>Splitting the data into Training data and Testing data</b>

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

<b>Model Training (Logistic Regression)<b/>

```python
model = LogisticRegression(max_iter=1000)
```

```python
# training the logistic regression model with training data
model.fit(X_train, Y_train)
```

<b>Model Evaluation</b><br/><br/>
Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
